- Añadir una columna “Years Playing”, que calcule el año actual menos la columna “Joined”

```python
from datetime import datetime

current_year = datetime.now().year

df['Years Playing'] = current_year - df['Joined']

df
```

- Generar un nuevo dataset que contenga el año (joined) y el número de jugadores (groupby)

```python
new_df = df["Joined"].value_counts().reset_index()

new_df.columns = ["Year Joined", "Number of Players"]

new_df = new_df.sort_values("Year Joined").reset_index(drop = True)

new_df
```

- Opcional: Generar un gráfico que contenga, por año, el número de jugadores

```python
import matplotlib.pyplot as plt

plt.bar(new_df['Year Joined'], new_df['Number of Players'])

plt.xlabel('Year Joined')
plt.ylabel('Number of Players')
plt.title('Number of Players Joined per Year')

plt.show()
```
